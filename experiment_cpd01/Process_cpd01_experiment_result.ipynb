```python
import csv
import pprint
```

```python
csv_folder = ""

# ここを変更してください。
csv_filename = "okada_cpd34-36_BDFMR"
#####

csv_file_path = csv_folder + csv_filename
```

```python
with open(csv_file_path) as f:
    reader = csv.reader(f)
    l = [row for row in reader]
```

```python
# 最初の3行は実験結果ではないため削除します。
only_results = l[4:]

only_results
```

```python
# フィルター、カットオフ周波数、アルファベットが一致しているものの正答率を表示します。
```

```python
# まずは、フィルター、カットオフ周波数、アルファベットを全て網羅したリストを作成します。
Filter = ['Butterworth Filter', 'Another Filter']

# ここを変更してください。
max_cpdx10 = 36
min_cpdx10 = 34
cpd_stridex10 = 1
#####

cpd_listx10 = [ cpd for cpd in range(min_cpdx10, max_cpdx10+1, cpd_stridex10)]

# ここを変更してください。
use_alphabets = ["B","D","F","M","R"]
#####

comprehensive_three_elements = []

for a in range(len(use_alphabets)):
    for f in range(len(Filter)):
        for c in range(len(cpd_listx10)):
            comprehensive_three_elements.append([Filter[f], use_alphabets[a], cpd_listx10[c]])

comprehensive_three_elements
```

```python
# 実験結果（only_results）をcomprehensive_three_elementsと比較できるように整形します。
shaped_results = [ [] for _ in range(len(only_results))]
for result_index in range(len(only_results)):
    # 先頭のスペースを削除して、Filter情報を入力します。
    shaped_results[result_index].append(only_results[result_index][1][1:])
     # 10倍して、cpd情報を追加します。
    shaped_results[result_index].append( int( float(only_results[result_index][2]) * 10 ) )
    # 正答情報を追加します。
    shaped_results[result_index].append( only_results[result_index][3][1:] )
    # 被験者の回答を追加します。
    shaped_results[result_index].append( only_results[result_index][4][1:] )

shaped_results
```

```python
# comprehensive_three_elementsに対応させた正答数のリストです。
comprehensive_correct_num = [ 0 for _ in range(len(comprehensive_three_elements)) ]

for result_index in range(len(shaped_results)):
    for comp_index in range(len(comprehensive_three_elements)):
        
        # フィルタが一致しているかを確認します。
        if shaped_results[result_index][0] ==  comprehensive_three_elements[comp_index][0]:
            
            # cpdが一致してるかを確認します。
            if shaped_results[result_index][1] ==  comprehensive_three_elements[comp_index][2]:
                
                # 正答アルファベットが一致してるかを確認します。
                if shaped_results[result_index][2] ==  comprehensive_three_elements[comp_index][1]:
            
                    # 被験者が正答しているかを判断します。
                    if shaped_results[result_index][2] == shaped_results[result_index][3]:
                        
                        comprehensive_correct_num[comp_index] += 1
                        
print(comprehensive_correct_num)
```

```python
# butterworth Filterと Another Filterを左右に並べる。タイトルは、「butterworth (Alphabet)」 or 「Another Filter (Alphabet)」。
# 一つ一つのグラフの中のx軸は、cpd。
# グラフのy軸は、正答数。

HowManyGraph = int( len(comprehensive_three_elements) / ( (len(cpd_listx10)) * 2 ) )

for graph_index in range(HowManyGraph):
    
    graph1_y = comprehensive_correct_num[ len(cpd_listx10) * (2*graph_index) : len(cpd_listx10) * (2*graph_index+1) ] 
    graph2_y = comprehensive_correct_num[ len(cpd_listx10) * (2*graph_index+1) : (len(cpd_listx10)) * (2*graph_index+2) ]
    
    # ここまではcpdは便宜上、10倍して扱っていました。なので、ここで1/10（実験した時の値に）してます。
    cpd_list = [ cpd_listx10[cpdx10_index]/10 for cpdx10_index in range(len(cpd_listx10)) ]
    
    # ここを変更してください。 
    # same_img_display_num と対応します。
    y_scale = [0,1,2]
    #####
    
    x_scale = cpd_list

    plt.title(comprehensive_three_elements[graph_index * (len(cpd_listx10)) * 2 ][1])
    plt.plot(cpd_list, graph1_y, marker="o", color = "red", linestyle = "--", label="Butterworth Filter")
    plt.plot(cpd_list, graph2_y, marker="v", color = "blue", linestyle = ":", label="Another Filter")
    plt.xlabel("cutoff cpd")
    plt.xticks(x_scale)
    plt.ylabel("subject correct num")
    plt.yticks(y_scale)
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0, fontsize=18)
    plt.show()
```
